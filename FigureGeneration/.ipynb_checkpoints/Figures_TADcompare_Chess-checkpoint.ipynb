```r
library(ggplot2)                                                                                                                                                                                                                       
library(viridis)                                                                                                                                                                                                                       
library(dplyr)                                                                                                                                                                                                                         
library(data.table)                                                                                                                                                                                                                    
                                                                                                                                                                                                                                       
## TAD compare                                                                                                                                                                                                                         
inPath <- "/Script/tadcompare/"                                                                                                                                                                                  
data <- fread(paste0(inPath, "feature_summary_true_expand_1mb.txt"))                                                                                                                                                                        
methodList = c("BandNorm", "3DVI", "Higashi", "scHiCluster")                                                     
```

```r
options(repr.plot.width = 20, repr.plot.height = 20, repr.plot.res = 100)
```

```r
data$method = factor(data$method, levels = methodList)
data %>% ggplot(aes(x = method, y = score * 100, fill = method)) +
 geom_violin() +
    geom_jitter(width = 0.2, color = "grey39") +
    geom_boxplot(width = 0.2) +
    scale_fill_viridis(discrete = TRUE, alpha = 0.4) +
    facet_grid(cell_type1 ~ cell_type2) +
    theme_bw(base_size = 25) +
    xlab("") +
    ylab("Concordance Percentage with Bulk Data") +
    ggpubr::rotate_x_text(angle = 20) +
    ggpubr::rremove("legend") +
    coord_cartesian(ylim = c(0, 100))
```

```r
data <- fread(paste0(inPath, "feature_summary_expand_1mb.txt"))  
data$method = factor(data$method, levels = methodList)

pdf("/Figures/TADcompare_clusterlabel.pdf", width = 20, height = 20)
data %>% ggplot(aes(x = method, y = score * 100, fill = method)) +
 geom_violin() +
    geom_jitter(width = 0.2, color = "grey39") +
    geom_boxplot(width = 0.2) +
    scale_fill_viridis(discrete = TRUE, alpha = 0.4) +
    facet_grid(cell_type1 ~ cell_type2) +
    theme_bw(base_size = 25) +
    xlab("") +
    ylab("Concordance Percentage with Bulk Data") +
    ggpubr::rotate_x_text(angle = 20) +
    ggpubr::rremove("legend") +
    coord_cartesian(ylim = c(0, 100))
dev.off()
```

```r
inPath = "/Script/chess/"
data <- fread(paste0(inPath, "feature_summary_true.txt"))                                                                                                                                                                        
cellList = c("GM12878", "H1ESC", "HAP1", "HFF", "IMR90")
```

```r
data$method = factor(data$method, levels = methodList)
data$cell_type1 = factor(data$cell_type1, levels = cellList)
data$cell_type2 = factor(data$cell_type2, levels = cellList)

index = which(data$cell_type1 == "HFF" & data$cell_type2 == "HAP1")
data$cell_type1[index] = "HAP1"
data$cell_type2[index] = "HFF"

pdf("/Figures/chess_truelabel.pdf", width = 20, height = 20)
data %>% ggplot(aes(x = method, y = score, fill = method)) +
 geom_violin() +
    geom_jitter(width = 0.2, color = "grey39") +
    geom_boxplot(width = 0.2) +
    scale_fill_viridis(discrete = TRUE, alpha = 0.4) +
    facet_grid(cell_type1 ~ cell_type2) +
    theme_bw(base_size = 25) +
    xlab("") +
    ylab("Correlation with Bulk Data") +
    ggpubr::rotate_x_text(angle = 20) +
    ggpubr::rremove("legend") 
dev.off()
```
