## This code is used to generate embeddings and UMAP for one-batch setting in the paper.

```r
library(data.table)
library(BandNorm)
schicluster = fread("/Ecker2019/human/schiclusterEmbedding1mb1batch/embed.txt")
infos = fread("/Ecker2019/human/cell_meta_cluster1Batch.txt",
             header = FALSE)
names = paste(infos$V1, ".txt", sep = "")
rownames(schicluster) = names
sums = fread("/paper_prep/Ecker2019_summary.txt")
sums = sums[sums$name %in% rownames(schicluster), ]

plot_embedding(schicluster, cell_info = sums[, c(1, 3)], 
               label = "Cell Type")
```

```r
hicdf = bandnorm(path = "/Ecker2019/human/190315_21yr", 
                 save = FALSE)
bandnormE = create_embedding(hic_df = hicdf, do_harmony = FALSE, chrs = paste0("chr", c(1:22, "X")), dim_pca = 1000)
sums = fread("/paper_prep/Ecker2019_summary.txt")
sums = sums[sums$name %in% rownames(bandnormE), ]

plot_embedding(bandnormE[, 1:20], cell_info = sums[, c(1, 3)], 
               label = "Cell Type")
```

```r
library(data.table)
library(BandNorm)
datas = "Lee2019"
summary = fread(paste0("/paper_prep/01072022/summaryFiles/",
                   datas, "_summary.txt"))

bandscale = fread(paste0("/paper_prep/01072022/embeddings/", 
              datas, "_BandScale.csv"))

sums = summary[grep(pattern = "190315_21yr", summary$name), ]
bandscale = bandscale[grep(pattern = "190315_21yr", summary$name), ]
rownames(bandscale) = sums$name

plot_embedding(bandscale, cell_info = sums[, c(1, 3)], 
               label = "Cell Type")
```
